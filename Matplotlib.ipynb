```python
t = np.linspace(0, 10, 51)
print(t)
```

```python
f = np.cos(t)
print(f)
```

```python
plt.plot(t, f, color='green')
plt.title('Graph f(t)')
plt.xlabel('Unit t')
plt.ylabel('Unit f')
plt.axis([0.5, 9.5, -2.5, 2.5])
plt.show()
```

```python
fig, ax = plt.subplots(nrows=2, ncols=2)
ax1, ax2, ax3, ax4 = ax.flatten()
ax1.plot(x, y1)
ax2.plot(x, y2)
ax3.plot(x, y3)
ax4.plot(x, y4)
ax1.set_title('Graph $y_1$')
ax2.set_title('Graph $y_2$')
ax3.set_title('Graph $y_3$')
ax4.set_title('Graph $y_4$')
ax1.set_xlim([-5, 5])
fig.set_size_inches(8, 6)
plt.subplots_adjust(wspace=0.3, hspace=0.3)
plt.show()
```
